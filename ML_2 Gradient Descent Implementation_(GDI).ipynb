```python
 # Task 1 :Implement Cost Function
def cost_func(y, y_pred):
    n = len(y)
    return (1/n) * np.sum((y - y_pred)**2)
```

```python
# Task 2 :Implement Gradient Descent
#weight initialization
w = 0.0          
b = 0.0
#LR
lr = 0.01
# #iterative updates
epochs = 100
costs = []

 for i in range(epochs):
    y_pred = w * x + b #current prediction
     # cost
    current_cost = cost_func(y, y_pred) 
    costs.append(current_cost)
    
    #  Gradients 
    n = len(x)
    #Gradient of weight
    dw = (-2 / n) * np.dot(x, (y - y_pred))
    #Gradient of bias
    db = (-2 / n) * np.sum(y - y_pred)
    
    #Update Step(Learning Rate )
    w = w - (lr * dw)
    b = b - (lr * db)
    
```

```python
print("Final Weight (w):", w)
print("Final Bias (b):  ", b)
print("Final MSE:  ", costs[-1])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(epochs), costs, color='green')
plt.title('COST (MSE) vs. ITERATIONS')
plt.xlabel('NUMBER of ITERATIONS')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```
